# Model Inference

## Objective

The objective of this notebook is to demonstrate how the trained Gradient Boosting model can be used to predict the sale price of new houses.

This notebook simulates the inference stage of a machine learning workflow, where unseen data is passed to the trained model to generate predictions.

```python
# ==========================
# Import Libraries
# ==========================

import joblib
import pandas as pd
```

```python
# ==========================
# Load Model
# ==========================

model = joblib.load("../models/tuned_gradient_boosting.pkl")
```

```python
# ==========================
# Loading Feature Names
# ==========================

X_train = joblib.load("../models/X_train_encoded.pkl")

training_columns = X_train.columns
```

```python
# ==========================
# Creating New House
# ==========================

new_house = {
    "MSSubClass": 20,
    "MSZoning": "RL",
    "LotFrontage": 80,
    "LotArea": 9600,
    "Street": "Pave",
    "Alley": "None",
    "LotShape": "Reg",
    "LandContour": "Lvl",
    "Utilities": "AllPub",
    "LotConfig": "Inside",
    "LandSlope": "Gtl",
    "Neighborhood": "NAmes",
    "Condition1": "Norm",
    "Condition2": "Norm",
    "BldgType": "1Fam",
    "HouseStyle": "1Story",
    "OverallQual": 7,
    "OverallCond": 5,
    "YearBuilt": 2005,
    "YearRemodAdd": 2008,
    "RoofStyle": "Gable",
    "RoofMatl": "CompShg",
    "Exterior1st": "VinylSd",
    "Exterior2nd": "VinylSd",
    "MasVnrType": "BrkFace",
    "MasVnrArea": 120,
    "ExterQual": "Gd",
    "ExterCond": "TA",
    "Foundation": "PConc",
    "BsmtQual": "Gd",
    "BsmtCond": "TA",
    "BsmtExposure": "No",
    "BsmtFinType1": "GLQ",
    "BsmtFinSF1": 700,
    "BsmtFinType2": "Unf",
    "BsmtFinSF2": 0,
    "BsmtUnfSF": 150,
    "TotalBsmtSF": 850,
    "Heating": "GasA",
    "HeatingQC": "Ex",
    "CentralAir": "Y",
    "Electrical": "SBrkr",
    "1stFlrSF": 850,
    "2ndFlrSF": 700,
    "GrLivArea": 1550,
    "BsmtFullBath": 1,
    "BsmtHalfBath": 0,
    "FullBath": 2,
    "HalfBath": 1,
    "BedroomAbvGr": 3,
    "KitchenAbvGr": 1,
    "KitchenQual": "Gd",
    "TotRmsAbvGrd": 7,
    "Fireplaces": 1,
    "FireplaceQu": "TA",
    "GarageType": "Attchd",
    "GarageYrBlt": 2005,
    "GarageFinish": "RFn",
    "GarageCars": 2,
    "GarageArea": 500,
    "GarageQual": "TA",
    "GarageCond": "TA",
    "PavedDrive": "Y",
    "WoodDeckSF": 100,
    "OpenPorchSF": 50,
    "EnclosedPorch": 0,
    "3SsnPorch": 0,
    "ScreenPorch": 0,
    "PoolArea": 0,
    "PoolQC": "None",
    "Fence": "None",
    "MiscFeature": "None",
    "MiscVal": 0,
    "MoSold": 6,
    "YrSold": 2010,
    "SaleType": "WD",
    "SaleCondition": "Normal"
}
```

```python
new_house["HouseAge"] = new_house["YrSold"] - new_house["YearBuilt"]

new_house["RemodelAge"] = (
    new_house["YrSold"] -
    new_house["YearRemodAdd"]
)

new_house["IsRemodeled"] = int(
    new_house["YearBuilt"] !=
    new_house["YearRemodAdd"]
)

new_house["TotalSF"] = (
    new_house["TotalBsmtSF"] +
    new_house["1stFlrSF"] +
    new_house["2ndFlrSF"]
)

new_house["TotalPorchSF"] = (
    new_house["OpenPorchSF"] +
    new_house["EnclosedPorch"] +
    new_house["3SsnPorch"] +
    new_house["ScreenPorch"]
)

new_house["OutdoorArea"] = (
    new_house["WoodDeckSF"] +
    new_house["OpenPorchSF"] +
    new_house["EnclosedPorch"] +
    new_house["3SsnPorch"] +
    new_house["ScreenPorch"]
)

new_house["TotalBathrooms"] = (
    new_house["FullBath"] +
    0.5 * new_house["HalfBath"]
)

new_house["HasGarage"] = int(new_house["GarageArea"] > 0)
new_house["HasBasement"] = int(new_house["TotalBsmtSF"] > 0)
new_house["HasPool"] = int(new_house["PoolArea"] > 0)
new_house["HasFireplace"] = int(new_house["Fireplaces"] > 0)
new_house["HasSecondFloor"] = int(new_house["2ndFlrSF"] > 0)
```

```python
# Converting to Dataframe

new_house_df = pd.DataFrame([new_house])
```

```python
# One Hot Encoding

new_house_df = pd.get_dummies(
    new_house_df,
    drop_first=True
)
```

```python
# Align Columns

new_house_df = new_house_df.reindex(
    columns=training_columns,
    fill_value=0
)
```

```python
# Prediction

prediction = model.predict(new_house_df)[0]

print(f"Predicted House Price: ${prediction:,.2f}")
```

```python
# Predict Multiple Houses

predictions = model.predict(new_house_df)

prediction_df = pd.DataFrame({
    "Predicted Price": predictions
})

prediction_df
```

```python
# Save Prediction

prediction_df.to_csv(
    "../reports/results/new_predictions.csv",
    index=False
)
```
