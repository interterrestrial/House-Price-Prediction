# Feature Engineering

## Objective

The objective of this notebook is to create meaningful new features from the existing Ames Housing dataset.

Feature engineering helps improve model performance by introducing domain-specific information that may not be directly available in the raw dataset.

In this notebook we will:

- Create time-based features
- Create area-based features
- Create bathroom-related features
- Create binary indicator features
- Create quality-based features
- Save the engineered dataset for preprocessing

```python
# Import Libraries

import pandas as pd 
import numpy as np 
```

```python
# Load Dataset
train_df = pd.read_csv("../data/raw/train.csv")
test_df = pd.read_csv("../data/raw/test.csv")
```

```python
# Create copies
train_fe = train_df.copy()
test_fe = test_df.copy()
```

```python
# ==========================================
# Time-Based Features
# ==========================================

# House Age
train_fe["HouseAge"] = train_fe["YrSold"] - train_fe["YearBuilt"]
test_fe["HouseAge"] = test_fe["YrSold"] - test_fe["YearBuilt"]

# Remodel Age
train_fe["RemodelAge"] = train_fe["YrSold"] - train_fe["YearRemodAdd"]
test_fe["RemodelAge"] = test_fe["YrSold"] - test_fe["YearRemodAdd"]

# Is Remodelled
train_fe["IsRemodeled"] = (
    train_fe["YearBuilt"] != train_fe["YearRemodAdd"]
).astype(int)

test_fe["IsRemodeled"] = (
    test_fe["YearBuilt"] != test_fe["YearRemodAdd"]
).astype(int)
```

```python
# ==========================================
# Area-Based Features
# ==========================================

# Total Square Feet
train_fe["TotalSF"] = (
    train_fe["TotalBsmtSF"] +
    train_fe["1stFlrSF"] +
    train_fe["2ndFlrSF"]
)

test_fe["TotalSF"] = (
    test_fe["TotalBsmtSF"] +
    test_fe["1stFlrSF"] +
    test_fe["2ndFlrSF"]
)

# Total Porch Area
train_fe["TotalPorchSF"] = (
    train_fe["OpenPorchSF"] +
    train_fe["EnclosedPorch"] +
    train_fe["3SsnPorch"] +
    train_fe["ScreenPorch"]
)

test_fe["TotalPorchSF"] = (
    test_fe["OpenPorchSF"] +
    test_fe["EnclosedPorch"] +
    test_fe["3SsnPorch"] +
    test_fe["ScreenPorch"]
)

# Outdoor Area
train_fe["OutdoorArea"] = (
    train_fe["WoodDeckSF"] +
    train_fe["OpenPorchSF"] +
    train_fe["EnclosedPorch"] +
    train_fe["3SsnPorch"] +
    train_fe["ScreenPorch"]
)

test_fe["OutdoorArea"] = (
    test_fe["WoodDeckSF"] +
    test_fe["OpenPorchSF"] +
    test_fe["EnclosedPorch"] +
    test_fe["3SsnPorch"] +
    test_fe["ScreenPorch"]
)
```

```python
# ==========================================
# Bathroom Features
# ==========================================

train_fe["TotalBathrooms"] = (
    train_fe["FullBath"] +
    (0.5 * train_fe["HalfBath"]) +
    train_fe["BsmtFullBath"] +
    (0.5 * train_fe["BsmtHalfBath"])
)

test_fe["TotalBathrooms"] = (
    test_fe["FullBath"] +
    (0.5 * test_fe["HalfBath"]) +
    test_fe["BsmtFullBath"] +
    (0.5 * test_fe["BsmtHalfBath"])
)
```

```python
# ==========================================
# Binary Features
# ==========================================

# Has Basement
train_fe["HasBasement"] = (train_fe["TotalBsmtSF"] > 0).astype(int)
test_fe["HasBasement"] = (test_fe["TotalBsmtSF"] > 0).astype(int)

# Has Garage
train_fe["HasGarage"] = (train_fe["GarageArea"] > 0).astype(int)
test_fe["HasGarage"] = (test_fe["GarageArea"] > 0).astype(int)

# Has Pool
train_fe["HasPool"] = (train_fe["PoolArea"] > 0).astype(int)
test_fe["HasPool"] = (test_fe["PoolArea"] > 0).astype(int)

# Has Fireplace
train_fe["HasFireplace"] = (train_fe["Fireplaces"] > 0).astype(int)
test_fe["HasFireplace"] = (test_fe["Fireplaces"] > 0).astype(int)

# Has Second Floor
train_fe["HasSecondFloor"] = (train_fe["2ndFlrSF"] > 0).astype(int)
test_fe["HasSecondFloor"] = (test_fe["2ndFlrSF"] > 0).astype(int)
```

```python
# ==========================================
# Check New Features
# ==========================================

new_features = [
    "HouseAge",
    "RemodelAge",
    "IsRemodeled",
    "TotalSF",
    "TotalPorchSF",
    "OutdoorArea",
    "TotalBathrooms",
    "HasBasement",
    "HasGarage",
    "HasPool",
    "HasFireplace",
    "HasSecondFloor"
]

train_fe[new_features].head()
```

```python
# ==========================================
# Save Engineered Dataset
# ==========================================

train_fe.to_csv(
    "../data/processed/train_feature_engineered.csv",
    index=False
)

test_fe.to_csv(
    "../data/processed/test_feature_engineered.csv",
    index=False
)
```
